```python
from collections import OrderedDict

class F:
    def __init__(self, x):
        self.x = x

class U:
    def __init__(self):
        self.Fs = {"1":F(1),
              "2":F(2)}

u = U();
for k, f in reversed(u.Fs.items()):
    f.x += 1

print([f.x for k, f in u.Fs.items()])
```

```python
for k, f in enumerate(reversed(u.Fs.values())):
    print(k)
    f.x += k

print([f.x for k, f in u.Fs.items()])
```
